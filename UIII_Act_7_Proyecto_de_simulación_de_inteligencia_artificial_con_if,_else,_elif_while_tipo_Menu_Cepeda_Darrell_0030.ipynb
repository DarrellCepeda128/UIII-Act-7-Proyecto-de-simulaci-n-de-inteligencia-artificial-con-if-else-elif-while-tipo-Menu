## Ejemplo electricista

```python
# --- ASISTENTE DE IA: MANTENIMIENTO ELÉCTRICO ---

print("==========================================")
print("   BIENVENIDO AL ASISTENTE IA DE OBRA   ")
print("==========================================")

# Variable para controlar el menú interactivo
opcion = 0

# Estructura repetitiva para mantener el programa activo
while opcion != 3:
    print("\n--- MENÚ DE MANTENIMIENTO ---")
    print("1. Diagnosticar estado de cables")
    print("2. Analizar consumo de energía")
    print("3. Salir del sistema")

    # Captura de datos interactiva
    opcion = int(input("\nSeleccione una opción (1-3): "))

    # Lógica de Inteligencia Artificial mediante condicionales
    if opcion == 1:
        print("\n--- DIAGNÓSTICO DE CABLES ---")
        voltaje = float(input("Ingrese el voltaje detectado (V): "))
        calibre = int(input("Ingrese el calibre del cable (AWG): "))
        dañado = input("¿El aislante se ve quemado? (si/no): ")

        # Razonamiento lógico basado en múltiples criterios
        if voltaje > 130 and calibre >= 14:
            print("IA DECISIÓN: ¡ALERTA! El cable es muy delgado para ese voltaje.")
            print("Sugerencia: Reemplace por un calibre 12 para evitar incendios.")
        elif dañado == "si" or voltaje > 240:
            print("IA DECISIÓN: RIESGO CRÍTICO.")
            print("Sugerencia: Corte el suministro inmediatamente; hay riesgo de arco eléctrico.")
        else:
            print("IA DECISIÓN: Los parámetros del cable son seguros.")

    elif opcion == 2:
        print("\n--- ANÁLISIS DE CONSUMO ---")
        watts = float(input("¿Cuántos Watts consume el aparato?: "))
        horas = float(input("¿Cuántas horas al día está encendido?: "))

        # Uso de operadores aritméticos para procesar datos
        consumo_diario = (watts * horas) / 1000

        if consumo_diario > 10:
            print("IA DECISIÓN: CONSUMO EXCESIVO DETECTADO.")
            print("Sugerencia: Este aparato representa un gasto alto. Considere mantenimiento preventivo.")
        elif consumo_diario > 0 and consumo_diario <= 10:
            print("IA DECISIÓN: Consumo dentro del rango eficiente.")
        else:
            print("IA DECISIÓN: El aparato parece estar apagado o desconectado.")

    elif opcion == 3:
        print("\nCerrando sesión del Asistente IA. ¡Buen trabajo en la obra!")

    else:
        # Caso para entradas no válidas
        print("Opción no válida. Por favor, elija entre 1 y 3.")

print("------------------------------------------")
print("Simulación finalizada.")

print("Programa realizado por Cepeda Darrell NL = 10")
```
